```python
id_model = "microsoft/Phi-3-mini-4k-instruct"
```

```python
model = AutoModelForCausalLM.from_pretrained(
    id_model,
    device_map = "cuda",
    torch_dtype = "auto",
    trust_remote_code = True,
    attn_implementation="eager"
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(id_model)
```

```python
pipe = pipeline("text-generation", model = model, tokenizer = tokenizer)
```

```python
generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.1, # 0.1 até 0.9
    "do_sample": True,
}
```

```python
prompt = "Explique o que é deep learning"
#prompt = "Quanto é 7 x 6 - 42?"

output = pipe(prompt, **generation_args)
```

```python
prompt = "Gere um código em python que escreva uma formula de algebra linear"

sys_prompt = "Você é um programador experiente. Retorne o código requisitado e forneça explicações breves se achar conveniente"

template = """<|system|>
{}<|end|>
<|user|>
"{}"<|end|>
<|assistant|>""".format(sys_prompt, prompt)

output = pipe(template, **generation_args)
print(output[0]['generated_text'])
```

```python
def gerar_formula_linear(coeficientes, constante):

    # Verifica se os coeficientes são válidos

    if not isinstance(coeficientes, list) or not all(isinstance(c, (int, float)) for c in coeficientes):

        raise ValueError("Coeficientes devem ser um lista de números.")

    if not isinstance(constante, (int, float)):

        raise ValueError("Constante deve ser um número.")


    # Formata a fórmula

    if len(coeficientes) == 1:

        return f"{coeficientes[0]}x = {constante}"

    else:

        formula = " + ".join(f"{c}x^{i}" for i, c in enumerate(coeficientes, start=1))

        return f"{formula} = {constante}"


# Exemplo de uso:

coeficientes = [3, -2, 1]  # Representa 3x^2 - 2x + 1

constante = 5

print(gerar_formula_linear(coeficientes, constante))
```
